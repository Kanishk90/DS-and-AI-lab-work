```python
#Subplots
plt.subplot(2,2,1)
x= np.linspace(1/np.e**2,np.e**3)
y = np.log(x)
plt.plot(x,y)
plt.title(r'$y = \ln(x)$')

plt.subplot(2,2,2)
x = np.linspace(0,3)
y = np.exp(x)
plt.plot(x,y)

plt.subplot(2,2,3)
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y)

plt.subplot(2,2,4)
x = np.linspace(0,2*np.pi)
y = np.tan(x)
plt.plot(x,y)
```
